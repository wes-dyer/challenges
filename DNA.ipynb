```python
#import module
import glob
```

```python
#list all files in working directory
print(glob.os.listdir())

#open dna_sequence txt file
dna = open('dna_sequence.txt').read().split('\n')

dna_range = dna[8:81] #generates a list from line 9 to 80 in the txt file. 
                      # removes the first 8 lines in the txt file (headers).

dna_list = "".join(dna_range) #links list together in one string
```

```python
# Define function
def dnamatching(sequence):
    '''
    A function that finds the complementary DNA sequence of a portion of DNA.
    
    Adenine-Thymine [A-T] & Cytosine-Guanine [C-G]
    '''
    
    matching_sequence = '' 
    
    #for loop 
    #finds and replaces the appropriate matching DNA nucleotide
    for base in sequence:
        
        # Guanine -> Cytosine 
        if base == 'G':
            matching_sequence += 'C'           # "+=" (example: A += B "adds" B to A by replacing the result in A) Concatenation.
            
        # Cytosine -> Guanine     
        elif base == 'C':
            matching_sequence += 'G'
            
        # Thymine -> Adenine     
        elif base == 'T':
            matching_sequence += 'A'
            
        # Adenine -> Thymine    
        elif base == 'A':
            matching_sequence += 'T'
            
         # Prompts error message if a DNA squence is not found   
        else:
            return print('Error! DNA sequence not found:', base)
            
    return matching_sequence
```

```python
# Call function
dna_matching_strand = dnamatching(dna_list)
```

```python
#prints matching sequence
print(dna_matching_strand)
```
