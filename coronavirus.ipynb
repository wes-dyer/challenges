```python
#import modules
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# retrieve data from github repository (URL links)

# total confirmed cases to date
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")

# total recovered cases to date
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv")

# total deaths to date 
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
```

```python
# Displays the first 5 rows of the dataset of confirmed cases.
confirmed.head() # calls the column headers
```

```python
# calculates the total number confirmed, recovered, and deaths

# sum of confirmed cases
total_confirmed = confirmed.iloc[:,4 :].sum()

# sum of recovered cases
total_recovered = recovered.iloc[:,4 :].sum()

# sum of deaths
total_deaths = deaths.iloc[:,4 :].sum()
```

```python
#data frame -> stores data into tables
confirmed_df = pd.DataFrame(confirmed)
recovered_df = pd.DataFrame(recovered)
deaths_df = pd.DataFrame(deaths)

#pd.to_datetime() -> convert argument to datetime
dates_confirmed = pd.to_datetime(confirmed_df.columns[4:])
dates_recovered = pd.to_datetime(recovered_df.columns[4:])
dates_deaths = pd.to_datetime(deaths_df.columns[4:])
```

```python
# Generate Figure

#figure size 
plt.figure(figsize = (10,8))

#plot each variable and customize for readability
plt.plot(dates_confirmed,total_confirmed,'yellow', linestyle ='--', label = 'Infected')
plt.plot(dates_recovered,total_recovered,'green', label = 'Recovered')
plt.plot(dates_deaths,total_deaths,'red', linestyle = ':',label = 'Deaths')

# more customization features
plt.xticks(rotation=45, fontsize = 14)
plt.yticks(fontsize = 14)
plt.ylabel('People', fontsize = 16)
plt.title('Coronavirus Timeseries', fontsize = 16)
plt.legend()
plt.show()
```
